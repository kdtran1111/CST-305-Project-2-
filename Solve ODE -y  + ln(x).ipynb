```python
#Kevin Tran


import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import time

start_time = time.time()
# Define the ODE as a function
def f(y, x):
    return -y + np.log(x)

# Initial conditions
y0 = 1.0
x0 = 2.0

# Create a time grid (x values) for integration
x_values = np.linspace(x0, 302, 1000)  # Adjust the end value and number of points as needed

# Solve the ODE using odeint
y_values = odeint(f, y0, x_values)

# Extract the solution
y_solution = y_values[:, 0]

# Print the result
for x, y in zip(x_values, y_solution):
    print(f"x = {x:.2f}, y = {y:.6f}")
    
# Plot the solution
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of dy/dx = -y + ln(x) using RKF45')
plt.grid(True)
plt.show()
    
end_time = time.time()

elapsed_time = end_time - start_time
print(f"Elapsed time: {elapsed_time} seconds")

```
